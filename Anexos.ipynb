# Imputación Multivariable Unilateral

## $ H = \frac{1}{m} \sum_{t=1}^{m}Q_t$

## $ V = \frac{1}{m} \sum_{t=1}^{m} V^2_t + (1 +\frac{1}{m})[\frac{1}{m-1}\sum_{t=1}^{m}(Q_t - H)^2]  $

## - Papel

```python
data_iterative_imp_papel = pd.DataFrame()
data_iterative_imp_papel["x_1"] = datos["Litros totales       (l) Fuentes Móviles"]
data_iterative_imp_papel["x_2"] = datos["Consumo agua(m3)"]
data_iterative_imp_papel["y"] = datos['Consumo de papel (resmas/mes)']
```

```python
imp_int_papel = IterativeImputer(random_state=0)
y_hat_papel = imp_int_papel.fit_transform(data_iterative_imp_papel)
```

```python
y_hat_papel= pd.DataFrame(y_hat_papel)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo (puedes reemplazar estos valores con tus propios datos)
categorias = ['Combustible', 'Papel', 'Energía', "Agua"]
valores = [0.47,
12.3,
21.4,
9.13
]

# Número de categorías
num_categorias = len(categorias)

# Ángulos para cada categoría
angulos = np.linspace(0, 2 * np.pi, num_categorias, endpoint=False).tolist()

# Añadir el primer elemento al final para cerrar el gráfico de araña
valores += valores[:1]
angulos += angulos[:1]

# Crear el gráfico de araña
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angulos, valores, color= "teal", alpha=0.55)

# Configuración de etiquetas
ax.set_thetagrids(np.degrees(angulos[:-1]), categorias)
ax.set_rlabel_position(180 / num_categorias)

# Añadir una etiqueta para cada punto
for i, valor in enumerate(valores[:-1]):
    ax.text(angulos[i], valor + 0.2, str(valor))

# Mostrar el gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo (puedes reemplazar estos valores con tus propios datos)
categorias = ['Combustible', 'Papel', 'Energía', "Agua"]

# Valores para tres conjuntos de datos
_2017 = [0.512454,
18.041847,
19.471948,
7.353427

]
_2018= [0.505376,
17.256121,
20.169612,
10.150359

]
_2019 = [0.521899,
13.314580,
20.180783,
11.233671

]
_2020 = [0.349726,
11.077729,
22.096480,
8.207632

]
_2021 = [0.402429,
6.674993,
21.304333,
8.936312

]
_2022= [0.525979,
7.426160,
25.141493,
8.877735
]

# Número de categorías
num_categorias = len(categorias)

# Ángulos para cada categoría
angulos = np.linspace(0, 2 * np.pi, num_categorias, endpoint=True).tolist()

# Crear el gráfico de araña con varios conjuntos de valores
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Trazar la primera serie de datos
ax.fill(angulos, _2017, color='b', alpha=0.25, label='2017')

# Trazar la segunda serie de datos
ax.fill(angulos, _2018, color='r', alpha=0.25, label='2018')

# Trazar la tercera serie de datos
ax.fill(angulos, _2019, color='g', alpha=0.25, label='2019')

ax.fill(angulos, _2020, color='y', alpha=0.25, label='2020')
ax.fill(angulos, _2021, color='purple', alpha=0.25, label='2021')
ax.fill(angulos, _2022, color='teal', alpha=0.25, label='2022')
# Configuración de etiquetas
ax.set_thetagrids(np.degrees(angulos), categorias)
ax.set_rlabel_position(180 / num_categorias)

# Añadir una leyenda
ax.legend()

# Mostrar el gráfico
plt.show()
```
